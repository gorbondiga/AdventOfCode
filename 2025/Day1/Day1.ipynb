```python
import polars as pl

input_file = 'input.txt'
df = pl.read_csv(input_file, has_header=False)

def convert_string(value):
    if "R" in value:
        return int(value[1:]) 
    else:
        return -int(value[1:])

result = []
operations = []
start = 50
prev = 0
for i, j in enumerate(df.iter_rows()):
    prev = start
    dir = 1 if j[0][0] == "R" else -1
    
    int_value = convert_string(j[0])    
    result.append(abs(int_value) // 100)
    int_value = abs(int_value) % 100 * dir
    start = start + int_value
    prev_hund = prev // 100
    curr_hund = start // 100
    hund_diff = abs(curr_hund - prev_hund)
    result.append(hund_diff)
    
    if dir < 0:
        if start % 100 == 0:
            result.append(1)
        if prev % 100 == 0:
            result.append(-1)
print("="* 8, "FInal result", "="*8)
print("Total score: ", sum(result))
```
